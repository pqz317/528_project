```python
%load_ext autoreload
%autoreload 2

import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
ve = scipy.io.loadmat("data/VE111.mat")
```

```python
ve.keys()
```

```python
# look at peak conductance stims, 
src = ve["phase1_scr_stim_peaks"][0, 0].squeeze()
```

```python
src_csr = ve["phase1_scr_stim_peaks"][0, 0].squeeze()
src_css = ve["phase1_scr_stim_peaks"][0, 1].squeeze()
src_csm = ve["phase1_scr_stim_peaks"][0, 2].squeeze()
fig, ax = plt.subplots()
ax.plot(src_csr, label="CSR")
ax.plot(src_css, label="CSS")
ax.plot(src_csm, label="CS-")
ax.set_xlabel("Times presented")
ax.set_ylabel("Skin conductance")
ax.legend()
```

```python
src_csr_all = ve["phase1_scr_stim_all"][0, 0].squeeze()
```
